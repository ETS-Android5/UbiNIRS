```python
import os
import numpy as np
import pandas as pd
```

```python
inputfolder = "./reference/raw_russian"
outputfolder = "./reference/processed_russian/"
all_filenames = os.listdir(inputfolder)
for filename in all_filenames:
    if filename.endswith(".csv"):
        input_fullpath = os.path.join(inputfolder, filename)
        df = pd.read_csv(input_fullpath, skiprows=21)
        df = df[["Wavelength (nm)", "Sample Signal (unitless)", "Reference Signal (unitless)"]]
        
        output_fullpath = os.path.join(outputfolder, filename)
        df.to_csv(output_fullpath, index=False, columns=None, header=None)
```
