```python
import os
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "app.settings")

import sys
sys.path.append("../")
sys.path.append("../../")

import django
django.setup()

import numpy as np
from django.db import models
from apps.glutentest.models import Wrap

all_samples = Wrap.objects.all()
```

```python
def sample_generator(label_language, samples_per_gluten_type):
    """Generator a sample sequence."""
    sample_subset_gluten = all_samples.filter(label_language=label_language, gluten_free_flag=False)
    sample_subset_gluten_free = all_samples.filter(label_language=label_language, gluten_free_flag=True)

    # Generate sequence as follows:
    # 1. For each visualization condition, pick ${samples_per_gluten_type} samples for gluten-free and non-gluten-free,
    #    respectively. ==> ${2 * samples_per_gluten_type} samples per visualization condition.
    # 2. Shuffle those samples above.
    # 3. Attach the shuffled sample list to the final list.
    # 4. Repeat 1~3 for all visualization methods. Note samples cannot be repeatably picked.
    current_label_samples = []
    for idx_method in range(len(visualization_methods)):
        current_method_samples = list(
            sample_subset_gluten[idx_method*samples_per_gluten_type:(idx_method+1)*samples_per_gluten_type]) + list(
            sample_subset_gluten_free[idx_method*samples_per_gluten_type:(idx_method+1)*samples_per_gluten_type])
        current_label_samples += list(np.random.permutation(current_method_samples))

    return current_label_samples
```

```python
# Generate sample sequences for all languages for all participants, respectively. Save as JSON.
all_sample_sequences = {}
num_participants = 30
label_languages = ["English", "Russian", "No label"]
visualization_methods = ["Framing", "Baseline", "Scale"]
samples_per_gluten_type = 3

for idx_participant in range(num_participants):
    participant = "P{:02d}".format(idx_participant+1)
    
    current_participant_sequences = {}
    for label_language in label_languages:
        current_participant_sequences[label_language] = sample_generator(label_language, samples_per_gluten_type)
    
    all_sample_sequences[participant] = current_participant_sequences
```
